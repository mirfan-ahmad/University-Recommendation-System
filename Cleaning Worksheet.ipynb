# Import Necessary Modules

```python
import pandas as pd
import openpyxl
```

# Loading workbook

```python
workbook = openpyxl.load_workbook('Institutes.xlsx')
worksheet = workbook['Institutes']
```

# Removing empty cells

```python
for row in worksheet.iter_rows():
    non_empty_cells = []
    for cell in row:
        if cell.value is not None:
            non_empty_cells.append(cell)
    for i, cell in enumerate(non_empty_cells):
        worksheet.cell(row=row[0].row+i, column=cell.column).value = cell.value
    for j in range(len(non_empty_cells), len(row)):
        worksheet.cell(row=row[0].row+j, column=row[j].column).value = None
workbook.save('Institutes.xlsx')
```
