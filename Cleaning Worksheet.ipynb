# Loading workbook

```python
df = pd.read_excel('Universities.xlsx')
df = df.fillna('-')
df
```

```python
def fetch(lst, start, end, row):
    if row.iloc[start] != '-':
        lst.append(row.iloc[start:end].tolist())

Universities = []
data = []
i = 0
for index, row in df.iterrows():
    if row.iloc[0] != '-':
        if i != 0:
            tmp = [bs, masters, mphill, phd, dip]
            Universities.append(tmp)
        data.append(row.iloc[0])
        bs = [row.iloc[1:5].tolist()]
        masters = [row.iloc[5:9].tolist()]
        mphill = [row.iloc[9:13].tolist()]
        phd = [row.iloc[13:17].tolist()]
        dip = [row.iloc[17:21].tolist()]
        i += 1
    else:
        fetch(bs, 1, 5, row)
        fetch(masters, 5, 9, row)
        fetch(mphill, 9, 13, row)
        fetch(phd, 13, 17, row)
        fetch(dip, 17, 21, row)
```

```python
DATA_FRAMES = []
for i in range(len(Universities)):
    df1 = pd.DataFrame(Universities[i][0], columns=['Bachelors - Programs', 'Year', 'Fee', 'Deadline'])
    df2 = pd.DataFrame(Universities[i][1], columns=['Masters - Programs', 'Year', 'Fee', 'Deadline'])
    df3 = pd.DataFrame(Universities[i][2], columns=['M-Phill - Programs', 'Year', 'Fee', 'Deadline'])
    df4 = pd.DataFrame(Universities[i][3], columns=['PHD - Programs', 'Year', 'Fee', 'Deadline'])
    df5 = pd.DataFrame(Universities[i][4], columns=['Diploma - Programs', 'Year', 'Fee', 'Deadline'])
    DATA_FRAMES.append([df1, df2, df3, df4, df5])
```
